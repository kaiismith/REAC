```python
import pandas as pd
```

```python
data = pd.read_csv('merged-data.csv')
data.head()
```

```python
def column_merge(column_tag1, column_tag2):
    column1 = data[column_tag1]
    column2 = data[column_tag2]
    return pd.concat([column1, column2])
    
```

```python
rightGaze = column_merge('Right Gaze 1', 'Right Gaze 2')
rightGaze = rightGaze.tolist()

leftGaze = column_merge('Left Gaze 1', 'Left Gaze 2')
leftGaze = leftGaze.tolist()

rightHeadpose = column_merge('Right HeadPose 1', 'Right HeadPose 2')
rightHeadpose = rightHeadpose.tolist()

leftHeadpose = column_merge('Left HeadPose 1', 'Left HeadPose 2')
leftHeadpose = leftHeadpose.tolist()
```
